```python
def Get_Drafted_Players(year):
    r = requests.get('https://www.basketball-reference.com/draft/NBA_{}.html'.format(year))
    html_text = r.text
    soup = BeautifulSoup(html_text,'lxml')

    table = soup.find(attrs={"id":'all_stats'})
    player_link_list = []
    body = table.find('tbody')
    for row in body.find_all('tr'):
        draft_pos = row.find('th')
        
        #Skip over the 2 lines in the Draft table that separate the first and second rounds
        if (draft_pos.text == 'Rk') | (draft_pos.text == ''):
            continue
            
        if row.find(attrs={'data-stat':'player'}) is None:
            continue
                 
        player = row.find(attrs={'data-stat':'player'})
        
        #Skip over any player who does not have a hyperlink in his name.  Likely didn't make an NBA start.
        if(player.find('a') is None):
            continue
            
        link = player.find('a')
        
        #Create a list of all the hyperlinks to the drafted players BBRef webpage
        player_link_list.append(link.get('href'))

    return player_link_list    
```

```python
def Extract_DataFrame(table_id):
    
    #Return None if the player does not have any tables in his webpage
    if(soup.find(attrs={"id":table_id}) is None):
       return None
    table = soup.find(attrs={"id":table_id})
    
    if table_id == 'all_per_game':
        dfs = pd.read_html(str(table))
        df = dfs[0]
        df.set_index('Season')

        for column in df:
            if not df[column].name.find("Unnamed"):
                del df[column]
    else:
        advanced_table = table.find(string=lambda text:isinstance(text,Comment))
        dfs = pd.read_html(str(advanced_table))
        df = dfs[0]
        df.set_index('Season')
        
        #Get rid of any of the Unnamed columns
        for column in df:
            if not df[column].name.find("Unnamed"):
                del df[column]
                
        #Get rid of the occasional problematic columns that are blank space or 0.1 as names   
        for column in df:     
            #print(df[column].name)
            #print(df.columns.get_loc(column))
            if (df[column].name == '\xa0.1') | (df[column].name == '\xa0'):
                #print('Entered If Statement')
                del df[column]
    
    return df;
```
